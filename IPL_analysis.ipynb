*italicized text*# Exploratory data analysis in python of IPL dataset (from 2008-2022)

[kaggle dataset ](https://www.kaggle.com/datasets/vora1011/ipl-2008-to-2021-all-match-dataset/data)




```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/IPL_Matches_2008_2022.csv')
print(df.head(5))
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
year2022 = df[df['Season']=="2022"]
print(year2022)
print(year2022.shape)
```

```python
year2022.value_counts('WinningTeam').plot(kind="bar",figsize=(14,6))
```

```python
year2022.value_counts('WinningTeam').plot(kind="pie",figsize=(14,6))
```
